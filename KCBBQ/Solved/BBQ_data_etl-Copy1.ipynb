```python
import pandas as pd
from sqlalchemy import create_engine
```

### Store CSV into DataFrame

```python
results_csv = "../Resources/results.csv"
competitions_csv = "../Resources/competitions.csv"
teams_csv = "../Resources/teams.csv"
```

```python
results_df = pd.read_csv(results_csv)
competitions_df = pd.read_csv(competitions_csv)
teams_df = pd.read_csv(teams_csv)
```

```python
results_df.head()
```

```python
competitions_df.head()
```

```python
teams_df.head()
```

```python
teams_df.rename(columns={"name": "team_name"}, inplace=True)
teams_df
```

```python
teams_df.rename(columns={"id": "team_id"}, inplace=True)
teams_df
```

```python
competitions_df.rename(columns={"kcbs_competition_id": "competition_id"}, inplace=True)
competitions_df
```

```python
bbq_join1 = pd.merge(teams_df, results_df, on="team_id")
bbq_join1.head()
```

```python
bbq_join2 = pd.merge(competitions_df, bbq_join1, on="competition_id")
bbq_join2.head()
```

```python
KC_BBQ = bbq_join2.loc[bbq_join2["location"] == "Kansas City, MO", :]
KC_BBQ.head()
```

```python
del KC_BBQ['created_at_y']
KC_BBQ.head()
```

```python
Kansas_City_BBQ = KC_BBQ[["team_name", "team_id", "place_rank", "category", "score", "percent_judges_certified", "name", "competition_id", "location", "state", "start_date"]]
Kansas_City_BBQ.head()
```
